```python
# remove both the leading and the trailing characters
Elon = {Text.strip() for Text in Elon.Text}
Elon
```

```python
# Joining the list into one string/text
join_text = ' '.join(Elon)
join_text
```

```python
# remove Twitter username handles from a given twitter text. (Removes @usernames)
from nltk.tokenize import TweetTokenizer

tokenizer = TweetTokenizer(strip_handles=True)
cl_Text = tokenizer.tokenize(join_text)
cl_Text
```

```python
Jo_Text = ' '.join(cl_Text)
Jo_Text
```

```python
#remove the punctuations means(, . / ; ' ' \)
import string # special operations on strings
no_punc_text = Jo_Text.translate(str.maketrans('', '', string.punctuation)) 
no_punc_text
```

```python
# remove https or url within text
import re
no_url_text=re.sub(r'http\S+', '', no_punc_text)
no_url_text
```

```python
from nltk.tokenize import word_tokenize

text_token = word_tokenize(no_url_text)
text_token
```

```python
len(text_token)
```

```python
# Remove Stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list = ['\x92','rt','ye','yeah','haha','Yes','U0001F923','I']
my_stop_words.extend(sw_list)

stop_tokens=[word for word in text_token if not word in my_stop_words]
print(stop_tokens)
```

```python
join_Text=' '.join(stop_tokens)
join_Text
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()

lemmas = []
for token in join_Text.split():
    lemmas.append(Lemmatizer.lemmatize(token))
    
len(lemmas)
```

# Feature Extaction
1. Using CountVectorizer

```python
# feature extraction


from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
```

```python
print(vectorizer.vocabulary_)
```

```python
print(vectorizer.get_feature_names()[100:200])
```

```python
print(X.toarray())
```

# 2. CountVectorizer with N-grams (Bigrams & Trigrams)

```python
vectorizer_ngram_range = CountVectorizer(analyzer='word',ngram_range=(1,1),max_features = 100)
bow_matrix_ngram =vectorizer_ngram_range.fit_transform(lemmas)
```

```python
print(vectorizer_ngram_range.get_feature_names())
```

```python
print(bow_matrix_ngram.toarray())
```

# 3. TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
max_features = TfidfVectorizer(norm='l2',analyzer='word',ngram_range=(1,3),max_features=500)
matix_ngram = max_features.fit_transform(lemmas)
```

```python
print(max_features.get_feature_names())
print(matix_ngram.toarray())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
```

```python
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    plt.figure(figsize=(40,30))
    plt.imshow(wordcloud)
    plt.axis('off')
    
# Generate Word Cloud

STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')
wordcloud=WordCloud(width=3000,height=2000,background_color='black',max_words=50,
                   colormap='Set1',stopwords=STOPWORDS).generate(join_Text)
plot_cloud(wordcloud)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sentiment_scores = sia.polarity_scores(join_Text)
```

```python
print(sentiment_scores)
```

# in percentage

```python
sentiment_percentages = {
    'negative': round(sentiment_scores['neg'] * 100, 2),
    'neutral': round(sentiment_scores['neu'] * 100, 2),
    'positive': round(sentiment_scores['pos'] * 100, 2),
    'compound': round(sentiment_scores['compound'] * 100, 2)
}
```

```python
for sentiment, percentage in sentiment_percentages.items():
    print(f'{sentiment.capitalize()}: {percentage}%')
```
